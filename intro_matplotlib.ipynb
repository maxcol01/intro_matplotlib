# Introduction to Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot()
plt.show()
```

```python
plt.plot([1,2,3,4]);
```

```python
x = [0, 1,2, 3,4,5]
y = [0, 1, 2, 3, 4, 5]
plt.plot(x,y)
plt.show()
```

```python
# 1st method (pyplot API)
fig = plt.figure() #create a figure
ax = fig.add_subplot()
ax.plot(x,y);
plt.show()
```

```python
# 2nd method (O-O API)
fig = plt.figure() # create a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x,y);
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x,y); # add somedata
type(fig), type(x)
```

## Matplotlib example workflow

```python
# 0. importing matplotlib (see above)
# 1. prepare some data:
xx = [1,2,3,4]
yy = [11, 22, 33, 44]

#2. setup plot
fig, ax = plt.subplots(figsize=(6,6)) #set the height and width
#3. plot the data
ax.plot(xx,yy)
#4. customize plot
ax.set(title = 'Simple plot', xlabel='x-axis', ylabel = 'y-axis')
#save & show 
fig.savefig('sample-plot.png')
```

## which one should be use ? Pytplot vs matplotlib OO method
* when plotting quickly, pyplot ok 
* but when plotting someting more adavance use OO

```python
over_50 = heart_disease[heart_disease.age > 50]
over_50
```

```python
#Pyplot method:
over_50.plot(kind='scatter', x='age', y = 'chol', c = 'target');
```

```python
# OO method mixed with Pytplot.
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age',y='chol',c='target',ax=ax);
ax.set_xlim([45,100]);
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))

# plot the data
scatter = ax.scatter(x=over_50.age, y=over_50.chol, c=over_50.target)

#customize

ax.set(title='Heart disease and cholesterol levels', xlabel='Age', ylabel='Cholesterol');

# add a legend

ax.legend(*scatter.legend_elements(), title='Target')

# add an horizontal line

ax.axhline(over_50.chol.mean(),linestyle='--')
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True);

# Add some data to ax0 and ax1
scatter1 = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target);
scatter2 = ax1.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target);
# Customize ax0 and ax1:
ax0.set(title='Heart disease and Cholesterol Levels', ylabel='Cholesterol');
ax1.set(title='Hear disease and levels', xlabel='Age', ylabel='Thalach');
#add legend to ax0 and ax1
ax0.legend(*scatter.legend_elements(), title='Target');
ax1.legend(*scatter.legend_elements(), title='Target');
# Add a horizontal line to axo and ax1
ax0.axhline(y=over_50.chol.mean(), linestyle='--');
ax1.axhline(y=over_50.thalach.mean(), linestyle='--');

#add a title to the figure:

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

#OO method from scratch
fig, ax = plt.subplots(figsize = (10,6));

# plot the data
scatter = ax.scatter(x=over_50.age, y=over_50.chol, c=over_50.target, cmap='winter');

#customize

ax.set(title='Heart disease and cholesterol levels', xlabel='Age', ylabel='Cholesterol');

# add a legend

ax.legend(*scatter.legend_elements(), title='Target');

# add an horizontal line

ax.axhline(over_50.chol.mean(),linestyle='--', c='red');
```

```python
# Customizing the x and y axes limitations:
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True);

# Add some data to ax0 and ax1
scatter1 = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target, cmap='winter');
scatter2 = ax1.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target, cmap='winter');
# Customize ax0 and ax1:
ax0.set(title='Heart disease and Cholesterol Levels', ylabel='Cholesterol');
ax1.set(title='Hear disease and levels', xlabel='Age', ylabel='Thalach');

ax0.set_xlim([50,80])
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
#add legend to ax0 and ax1
ax0.legend(*scatter.legend_elements(), title='Target');
ax1.legend(*scatter.legend_elements(), title='Target');
# Add a horizontal line to axo and ax1
ax0.axhline(y=over_50.chol.mean(), linestyle='--', c='red');
ax1.axhline(y=over_50.thalach.mean(), linestyle='--',c='red');

#add a title to the figure:

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
